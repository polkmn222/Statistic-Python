### * Numpy 복습

```python
import numpy as np
```

```python
a = np.array([1,2,3])
print(type(a))
print(a.shape) # 3인 이유는 tuple의 법칙을 위반하지 않기 위해.
print('a 자료는 {0}차원'.format(a.ndim))
```

```python
# 2차원 데이터 살펴보기- 복습
array2d = np.array([[1,2,3],
                    [11,22,33]])
print('array2d type:', type(array2d))
print('array2d shape:', array2d.shape)
print('array2d 차원:', array2d.ndim)

# 3차원 데이터 살펴보기- 복습
array3d = np.array([[[1,2,3],
                    [11,22,33]]])
print('array3d type:', type(array3d))
print('array3d shape:', array3d.shape)
print('array3d 차원:', array3d.ndim)
```

```python
## Numpy의 기본 필요 함수
# ndarray를 쉽게 만들 수 있는 - arange, ones, zeros
sq_array = np.arange(4)
print(sq_array)
print(sq_array.dtype, sq_array.shape)
```

```python
# 1로 채워진 데이터 - ones
one_array = np.ones((2,3), dtype='int32')
print(one_array)
print(one_array.dtype, one_array.shape)
```

```python
# 0으로 채워진 데이터 - zeros
z_array = np.zeros((2,3), dtype='int32')
print(z_array)
print(z_array.dtype, z_array.shape)
```

```python
### ndarray의 초기화

## np.full() ndarray에서 사용자가 지정한 값을 삽입
full_array = np.full((3,2,1),7) # 상수값 7 앞의 수는 shape을 넣을 수 있다.
full_array

## np.eye() 대각선으로 1이고 나머지는 0인 2차원행렬 - 희소행렬(sparse)

eye_array = np.eye(4)
eye_array

## np.random.random() 임의의 값을 가지는 array생성이 가능
np.random.random((2,2)) # 0~1사이의 임의의 값으로 채워진 배열 생성
```

```python
### Numpy의 연산
## 그냥 손쉽게 사칙연산을 하셔도 좋지만
## np.add(), np.subtract(), np.multiply(), np.divide()

a = np.array([1,2,3])
b = np.array([11,22,33])

## 함수를 통한 덧셈
np.add(a,b)

## 함수를 통한 뺄셈 :: b에서 a를 뺍니다.
np.subtract(a,b)

## 함수를 통한 곱셈 
# 자동완성은 ctrl + space bar를 통해서 colab에서는 이뤄집니다. 
np.multiply(a,b)

## 함수를 통한 나눗셈 
# 자동완성은 ctrl + space bar를 통해서 colab에서는 이뤄집니다. 
np.divide(a,b)
```
